```
df = (spark.read.format("csv")
  .option("inferSchema", "true")
  .option("header", "true")
  .option("sep", ";")
  .load("/FileStore/tables/arquivo_geral.csv")
  .createOrReplaceTempView("tabela_covid")
     )
```

```
%sql
SELECT * FROM tabela_covid
```

```
%sql
SELECT regiao, estado, casosNovos FROM tabela_covid
```

Fazer a contagem de casos novos por Região

```
%sql
SELECT regiao, COUNT(casosNovos) AS QTD_CASOS_NOVOS FROM tabela_covid GROUP BY regiao
```

Fazer a contagem de casos novos por Região e mostrar apenas onde a quantidade for mairo que 500

```
%sql
SELECT regiao, COUNT(casosNovos) AS QTD_CASOS_NOVOS FROM tabela_covid GROUP BY regiao HAVING QTD_CASOS_NOVOS > 500
```

Fazer a contagem de casos novos da Região Nordeste

```
%sql
SELECT regiao, COUNT(casosNovos) AS QTD_CASOS_NOVOS FROM tabela_covid WHERE regiao == 'Nordeste' GROUP BY regiao
```
