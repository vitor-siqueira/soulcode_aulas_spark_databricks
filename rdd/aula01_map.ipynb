```
lista_nomes = ['Vitor', 'Jéssica', 'Giovanna', 'Théo', 'Miriam', 'Santinho', 'Scooby', 'Nome1', 'Nome2', 'Nome3']

rdd1 = spark.sparkContext.parallelize(lista_nomes)

rdd1.collect()

rdd1.map(lambda x: x.upper()).collect()
```

```
lista_nomes = ["SoulCode Curso de Engenharia de Dados", "SoulCode Curso de Engenharia de Dados", "Professores Bismark William", "Igor Gondim"]

rdd3 = spark.sparkContext.parallelize(lista_nomes)

rdd3.collect()

rdd3.flatMap(lambda y: y.split(" ")).map(lambda z: z.upper()).collect()
```

```
rdd3.flatMap(lambda z: z.upper()).collect()
```

```
rdd4 = spark.sparkContext.parallelize(range(1, 51))

rdd4.filter(lambda x: x <= 30).collect()
```
